```python
import json

def generate_cw_json(cw):
    cw_json = {
      cw: {
        cw: {
          f"./Adams res {cw} 160": {
            "cmd": f"nohup ./Adams res {cw} 160 >> {cw}.out 2>&1 &"
          },
          f"./Adams prod_mod {cw} S0 160": {
            "priority": 99,
            "cmd": f"nohup ./Adams prod_mod {cw} S0 160 >> {cw}_prod.out 2>&1 &",
            "pre": [
              f"./Adams res {cw} 160"
            ]
          },
          f"./Adams export_mod {cw} S0 160": {
            "cmd": f"./Adams export_mod {cw} S0 160",
            "pre": [
              f"./Adams prod_mod {cw} S0 160"
            ]
          },
          f"cp {cw}_AdamsSS.db AdamsSS/": {
            "cmd": f"cp {cw}_AdamsSS.db AdamsSS/",
            "pre": [
              "mkdir AdamsSS",
              f"./Adams export_mod {cw} S0 160"
            ]
          }
        }
      }
    }
    return cw_json

def generate_map_json(key, cw1, cw2):
    key1 = key.replace("__", " ")
    map_json = {
        key: {
            "priority": 99,
            "./Adams map_coh {}".format(key1): {
                "cmd": "./Adams map_coh {}".format(key1)
            },
            "./Adams map_res {} 1160".format(key1): {
                "cmd": "nohup ./Adams map_res {} 1160 >> map_{}.out 2>&1 &".format(key1, key),
                "pre": [
                "./Adams map_coh {}".format(key1),
                "./Adams res {} 160".format(cw1),
                "./Adams res {} 160".format(cw2)
                ]
            },
            "./Adams export_map {} 1160".format(key1): {
                "cmd": "./Adams export_map {} 1160".format(key1),
                "pre": [
                "./Adams export_mod {} S0 160".format(cw1) if cw1 != "S0" else "./Adams export S0 160",
                "./Adams export_mod {} S0 160".format(cw2) if cw2 != "S0" else "./Adams export S0 160",
                "./Adams map_res {} 1160".format(key1)
                ]
            },
            "cp map_AdamsSS_{}.db AdamsSS/".format(key): {
                "cmd": "cp map_AdamsSS_{}.db AdamsSS/".format(key),
                "pre": [
                "mkdir AdamsSS",
                "./Adams export_map {} 1160".format(key1)
                ]
            }
        }
    }
    return map_json

def remove_comments(content: str):
    return "\n".join(line.split("//")[0] for line in content.split("\n"))

with open("Adams.json", "r") as f:
    json_str = f.read()
    js = json.loads(remove_comments(json_str))
```

```python
maps_json = {}
for key, value in js["maps"].items():
    maps_json |= generate_map_json(key, value["from"], value["to"])
print(json.dumps(maps_json, indent=2))
```
